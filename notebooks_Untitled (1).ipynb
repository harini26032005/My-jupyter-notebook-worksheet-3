```python
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))
if num1 > num2:
    print("The first number is larger.")
elif num2 > num1:
    print("The second number is larger.")
else:
    print("Equal")
```

```python
name = input("Enter your name: ")
marks = int(input("Enter the marks: "))
print("Entered marks are:", marks)
if 100 >= marks >= 85:
    print("Excellent,",name,"You have A pass")
elif marks >= 75:
    print("Great,", name, "You have B pass")
elif marks >= 65:
    print("Good,", name, "You have C pass")
elif marks >= 50:
    print("Fair,", name, "You have D pass")
else:
    print("Sorry,", name, "You have failed. Try again")
```
